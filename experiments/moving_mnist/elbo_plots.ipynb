```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

# Batch losses

```python
rb_batch_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_batch_losses.npy')
rebar_batch_losses = np.load('./mnist_vae_results/moving_mnist_vae_rebar_batch_losses.npy')
```

```python
nvil_batch_losses = np.load('./mnist_vae_results/moving_mnist_vae_nvil_batch_losses.npy')
```

```python
reinforce_double_bs = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_double_bs_batch_losses.npy')
```

```python
plt.plot(rb_batch_losses, 'r+-')
plt.plot(rebar_batch_losses, 'b+-')
plt.plot(nvil_batch_losses, 'c+-')
plt.plot(reinforce_double_bs, 'y+-')

plt.ylabel('kl')
plt.xlabel('epoch')
```

```python
rb_batch_timing = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_batch_timing.npy')
rebar_batch_timing = np.load('./mnist_vae_results/moving_mnist_vae_rebar_batch_timing.npy')
nvil_batch_timing = np.load('./mnist_vae_results/moving_mnist_vae_nvil_batch_timing.npy')
```

```python
plt.plot(rb_batch_timing.cumsum(), rb_batch_losses, 'k+-')
plt.ylabel('kl')
plt.ylabel('seconds')
```

```python
rb_train_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_train_losses.npy')
rb_test_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_test_losses.npy')
rb_test_timing = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_test_timing.npy')
```

```python
plt.plot(rb_test_timing - rb_test_timing[0], rb_train_losses, 'k+--')
plt.plot(rb_test_timing - rb_test_timing[0], rb_test_losses, 'k+-')
```
