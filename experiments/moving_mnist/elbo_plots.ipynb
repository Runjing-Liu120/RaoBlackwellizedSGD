```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

# Batch losses

```python
rb_batch_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_batch_losses.npy')
rebar_batch_losses = np.load('./mnist_vae_results/moving_mnist_vae_rebar_batch_losses.npy')
```

# Test losses

```python
rb_test_timing = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_test_timing.npy')
rb_test_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_test_losses.npy')
```

```python
reinforce_bl_test_timing = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_double_bs_test_timing.npy')
reinforce_bl_test_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_double_bs_test_losses.npy')
```

```python
rebar_test_timing = np.load('./mnist_vae_results/moving_mnist_vae_rebar_test_timing.npy')
rebar_test_losses = np.load('./mnist_vae_results/moving_mnist_vae_rebar_test_losses.npy')
```

```python
nvil_test_timing = np.load('./mnist_vae_results/moving_mnist_vae_nvil_test_timing.npy')
nvil_test_losses = np.load('./mnist_vae_results/moving_mnist_vae_nvil_test_losses.npy')
```

```python
# plot for paper 
fig = plt.figure()
ax = plt.subplot(111)

ax.plot(rb_test_timing - rb_test_timing[0], 
         np.log(rb_test_losses), 'r-+', label = 'rb5')

ax.plot(reinforce_bl_test_timing - reinforce_bl_test_timing[0], 
         np.log(reinforce_bl_test_losses), 'y-+', label = 'reinforce+bs')

ax.plot(rebar_test_timing - rebar_test_timing[0], 
         np.log(rebar_test_losses), 'b-+', label = 'rebar')

ax.plot(nvil_test_timing - nvil_test_timing[0], 
         np.log(nvil_test_losses), 'c-+', label = 'nvil')

plt.xlabel('seconds', fontsize = 16)
plt.ylabel('Negative ELBO', fontsize = 16)
```

```python
rb_train_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_train_losses.npy')
rb_test_losses = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_test_losses.npy')
rb_test_timing = np.load('./mnist_vae_results/moving_mnist_vae_reinforce_doublebs_topk5_test_timing.npy')
```
