```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline  
```

# train losses

```python
# fully marginallized
batch_timing_fully_marg = np.load('./mnist_vae_results/ss_vae_fully_marg_batch_timing.npy')
batch_losses_fully_marg = np.load('./mnist_vae_results/ss_vae_fully_marg_batch_losses.npy')

np.mean(batch_timing_fully_marg[1:])
```

```python
# rebar 
batch_timing_rebar = np.load('./mnist_vae_results/ss_vae_rebar_batch_timing.npy')
batch_losses_rebar = np.load('./mnist_vae_results/ss_vae_rebar_batch_losses.npy')

np.mean(batch_timing_rebar[1:])
```

```python
# nvil 
batch_timing_nvil = np.load('./mnist_vae_results/ss_vae_nvil_batch_timing.npy')
batch_losses_nvil = np.load('./mnist_vae_results/ss_vae_nvil_batch_losses.npy')

np.mean(batch_timing_nvil[1:])
```

```python
# gumbel 
batch_timing_gumbel = np.load('./mnist_vae_results/ss_vae_gumbel_batch_timing.npy')
batch_losses_gumbel = np.load('./mnist_vae_results/ss_vae_gumbel_batch_losses.npy')

np.mean(batch_timing_gumbel[1:])
```

```python
start_epoch = 0
```

```python
# plot across epoch
plt.plot(batch_losses_fully_marg[start_epoch:-1], 'k:+')

plt.plot(batch_losses_reinforce[start_epoch:-1], 'y:+')

plt.plot(batch_losses_rebar[start_epoch:-1], 'b:+')

plt.plot(batch_losses_reinforce_topk1[start_epoch:-1], 'r-+')

plt.plot(batch_losses_nvil[start_epoch:-1], 'c-+')

plt.plot(batch_losses_gumbel[start_epoch:-1], 'g-+')

plt.xlabel('epoch')
plt.ylabel('negative elbo')
```

```python
plt.plot(batch_timing_fully_marg.cumsum()[start_epoch:-1], 
         batch_losses_fully_marg[start_epoch:-1], 'k:')

plt.plot(batch_timing_reinforce.cumsum()[start_epoch:-1], 
         batch_losses_reinforce[start_epoch:-1], 'g:')

plt.plot(batch_timing_rebar.cumsum()[start_epoch:-1], 
         batch_losses_rebar[start_epoch:-1], 'b:')

plt.plot(batch_timing_reinforce_topk1.cumsum()[start_epoch:-1], 
         batch_losses_reinforce_topk1[start_epoch:-1], 'r:')

plt.plot(batch_timing_nvil.cumsum()[start_epoch:-1], 
         batch_losses_nvil[start_epoch:-1], 'c:')

plt.plot(batch_timing_gumbel.cumsum()[start_epoch:-1], 
         batch_losses_gumbel[start_epoch:-1], 'g--')
```

```python
# plt.plot(test_timing_gumbel - test_timing_gumbel[0], 
#          train_losses_gumbel, 'g:+')
plt.plot(test_timing_gumbel - test_timing_gumbel[0], 
         test_losses_gumbel, 'g-+')

# plt.plot(test_timing_nvil - test_timing_nvil[0], 
#          train_losses_nvil, 'c:+')
plt.plot(test_timing_nvil - test_timing_nvil[0], 
         test_losses_nvil, 'c-+')

# plt.plot(test_timing_rebar - test_timing_rebar[0], 
#          train_losses_rebar, 'b:+')
plt.plot(test_timing_rebar - test_timing_rebar[0], 
         test_losses_rebar, 'b-+')


# plt.plot(test_timing_reinforce_topk1 - test_timing_reinforce_topk1[0], 
#          train_losses_reinforce_topk1, 'r:+')
plt.plot(test_timing_reinforce_topk1 - test_timing_reinforce_topk1[0], 
         test_losses_reinforce_topk1, 'r-+')

# plt.plot(test_timing_reinforce - test_timing_reinforce[0], 
#          train_losses_reinforce, 'y:+')
plt.plot(test_timing_reinforce - test_timing_reinforce[0], 
         test_losses_reinforce, 'y-+')

# plt.plot(test_timing_fully_marg - test_timing_fully_marg[0], 
#          train_losses_fully_marg, 'k:+')
plt.plot(test_timing_fully_marg - test_timing_fully_marg[0], 
         test_losses_fully_marg, 'k-+')

plt.xlabel('seconds', fontsize = 16)
plt.ylabel('Negative ELBO', fontsize = 16)
```

```python
# plot for paper 

fig = plt.figure()
ax = plt.subplot(111)
ax.plot(test_timing_fully_marg - test_timing_fully_marg[0], 
         test_losses_fully_marg, 'k-+', label = 'True grad')

ax.plot(test_timing_reinforce_topk1 - test_timing_reinforce_topk1[0], 
         test_losses_reinforce_topk1, 'r-+', label = 'RB1-REINFORCE')

ax.plot(test_timing_reinforce - test_timing_reinforce[0], 
         test_losses_reinforce, 'y-+', label = 'REINFORCE')


ax.plot(test_timing_gumbel - test_timing_gumbel[0], 
         test_losses_gumbel, 'g-+', label = 'Gumbel-Softmax')

ax.plot(test_timing_nvil - test_timing_nvil[0], 
         test_losses_nvil, 'c-+', label = 'NVIL')

ax.plot(test_timing_rebar - test_timing_rebar[0], 
         test_losses_rebar, 'b-+', label = 'REBAR')


plt.xlabel('seconds', fontsize = 16)
plt.ylabel('Negative ELBO', fontsize = 16)


plt.xlim((0, 2500))
plt.ylim((97, 109))

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('../icml_figures/ss_mnist_elbo_path.png')
```
