```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
```

# look at it across time

```python
# fully marginallized
batch_timing_fully_marg = np.load('./mnist_vae_results/ss_vae_fully_marg_batch_timing.npy')
batch_losses_fully_marg = np.load('./mnist_vae_results/ss_vae_fully_marg_batch_losses.npy')
```

```python
# reinforce
batch_timing_reinforce = np.load('./mnist_vae_results/ss_vae_reinforce_batch_timing.npy')
batch_losses_reinforce = np.load('./mnist_vae_results/ss_vae_reinforce_batch_losses.npy')
```

```python
# rb reinforce
batch_timing_reinforce_topk1 = np.load('./mnist_vae_results/ss_vae_reinforce_topk1_batch_timing.npy')
batch_losses_reinforce_topk1 = np.load('./mnist_vae_results/ss_vae_reinforce_topk1_batch_losses.npy')
```

```python
# rebar 
batch_timing_rebar = np.load('./mnist_vae_results/ss_vae_rebar_batch_timing.npy')
batch_losses_rebar = np.load('./mnist_vae_results/ss_vae_rebar_batch_losses.npy')
```

```python
start_epoch = 0
```

```python
plt.plot(batch_timing_fully_marg.cumsum()[start_epoch:-1], 
         batch_losses_fully_marg[start_epoch:-1], 'k:+')

plt.plot(batch_timing_reinforce.cumsum()[start_epoch:-1], 
         batch_losses_reinforce[start_epoch:-1], 'g:+')

plt.plot(batch_timing_rebar.cumsum()[start_epoch:-1], 
         batch_losses_rebar[start_epoch:-1], 'b:+')


plt.plot(batch_timing_reinforce_topk1.cumsum()[start_epoch:-1], 
         batch_losses_reinforce_topk1[start_epoch:-1], 'r-+')

```
